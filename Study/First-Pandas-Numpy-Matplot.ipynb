```
# DataFrame
values = [[1,2,3],[4,5,6],[7,8,9]]
index = ['one','two','three']
columns = ['A','B','C']
df = pd.DataFrame(values, index=index, columns=columns)
print(df.index)
print(df.columns)
print(df.values)
print(df)
```

```
# Create DataFrame from list
data = [
  ['1000', 'Steve', 90.70],
  ['1001', 'James', 78.09],
  ['1002', 'Doyeon', 97.12],
  ['1003', 'Jane', 45.44],
  ['1004', 'Pilwoong', 68.12],
  ['1005', 'Tony', 88.82]
]
df = pd.DataFrame(data)
print(df)
```

```
# Create DataFrame from dictionary
columns = ["num", "name", "grade"]
datas = [['1000','1010','1020','1030','1040','1050'],
         ['Steve', 'James', 'Doyeon', 'Jane', 'Pilwoong', 'Tony'],
         [10, 30, 40, 30, 20, 50]
]
data = dict(zip(columns, datas))
df = pd.DataFrame(data)
print(df)
```
